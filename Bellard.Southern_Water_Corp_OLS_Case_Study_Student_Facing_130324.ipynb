Now a quick analysis might show-case there are many *outliers* that exist in our dataset.

As we all know, Outliers can easily *skew* our analysis and we might want to remove them.

Let's experiment with removing Outliers with the following steps:

The general logic for removing an outliers is:

<b> Any Value that is either Q1 - 1.5x IQR or greater than Q3 + 1.5x IQR is considered an outlier. </b>

Where Q1 = Quartile 1 and Q3 = Quartile 3

Let's break these down to a few simple steps and tackle how to calculate this.

<b>

    1. Define the Quartiles Q1 and Q3. 
    
    2. Calculate the Interquartile Range (Q3 - Q1)
    
    3. Create two new variables; Lower_Range (Q1 - 1.5x IQR) and Upper_Range ( Q3 + 1.5x IQR).
    
    4. Using Boolean Subsetting, filter the Dataframe for Outliers and remove them
    
    5. Calculate what the proportion of Outliers exist (i.e. Number of entries left after Outlier removal / Number of total entries in dataset) 

</b>

### Step 6: <span style="color:green">Identify Outliers</span> 

Continuing in from Step 5 - we still need to calculate how we can remove our outliers.

We're aware of the overall formula as shown per below.

<b> Outlier = Q1 - 1.5x IQR OR Q3 + 1.5x IQR </b>

Now let's work out how we can identify these Outliers step by step. 

<b> i) Define two new variables, Lower_Limit and Upper_Limit can be calculated as:</b>
<p>
<li> Lower_Limit = Q1 - 1.5 * IQR </li>
<li> Upper_Limit = Q3 + 1.5 * IQR </li> 
   
<b> ii) Using Lower_Limit and Upper_Limit with an OR (|) condition, filter the dataframe to include *only* the Outliers. How many Outliers do we have?</b>
    
The syntax should look like this:
    
<b>`Outliers = dataframe_raw[((dataframe_raw[IQR.index] < lower_limit) | ((dataframe_raw[IQR.index] > upper_limit))).any(axis=1)]`</b>
    
The `.any(axis=1)` indicates that for *any* entry in the columns that matches the criteria of being *less* than the lower limit or *greater* than the upper limit, for that column (axis=1), that row will be removed. 

If you're stuck, <a href = https://stackoverflow.com/questions/54398554/how-to-remove-outliers-in-python> here's </a> a link that will show you how you can remove Outliers in Python! 

<b> iii) What is the percentage of data that remains after we've removed the Outliers from the dataframe_raw dataset?</b>
    

## Please put your code here

```python


# Assuming dataframe_raw is already defined and contains the necessary data
# Convert all columns to numeric, forcing errors to NaN
dataframe_raw = dataframe_raw.apply(pd.to_numeric, errors='coerce')

# Calculate Q1 (25th percentile) for all columns
Q1 = dataframe_raw.quantile(0.25)

# Calculate Q3 (75th percentile) for all columns
Q3 = dataframe_raw.quantile(0.75)

# Calculate the Interquartile Range (IQR) for all columns
IQR = Q3 - Q1

# Define Lower_Limit and Upper_Limit
Lower_Limit = Q1 - 1.5 * IQR
Upper_Limit = Q3 + 1.5 * IQR

# Identify the outliers based on the Lower_Limit and Upper_Limit
Outliers = dataframe_raw[((dataframe_raw[IQR.index] < Lower_Limit) | (dataframe_raw[IQR.index] > Upper_Limit)).any(axis=1)]

# Count the number of outliers
num_outliers = Outliers.shape[0]

# Calculate the number of rows in the original dataframe
total_rows = dataframe_raw.shape[0]

# Calculate the number of rows after removing the outliers
rows_after_removal = total_rows - num_outliers

# Calculate the percentage of data that remains
remaining_percentage = (rows_after_removal / total_rows) * 100

# Print the results
print(f"Number of Outliers: {num_outliers}")
print(f"Percentage of data that remains after removing outliers: {remaining_percentage:.2f}%")

```

### Step 7: <span style="color:green">Create a Boxplot without Outliers</span> 

i) Create a new dataframe called no_outliers and using the ~ operator, remove all the Outliers from the dataframe <p>
ii) Using the no_outliers dataframe, create two box plots as per below: <p>
    <li> A boxplot when PUMP FAILURE is 1 (Failure) </li>
    <li> A boxplot when PUMP FAILURE is 0 (Normal Behaviour) </li>

## Please put your code here

```python



# Assuming Q1, Q3, and IQR have already been calculated from dataframe_raw
# Define the lower and upper limits for outlier detection
Lower_Limit = Q1 - 1.5 * IQR
Upper_Limit = Q3 + 1.5 * IQR

# Identify rows with outliers in the numeric columns
Outliers = dataframe_raw[((dataframe_raw[IQR.index] < Lower_Limit) | (dataframe_raw[IQR.index] > Upper_Limit)).any(axis=1)]

# Remove outliers to create a new dataframe
no_outliers = dataframe_raw[~((dataframe_raw[IQR.index] < Lower_Limit) | (dataframe_raw[IQR.index] > Upper_Limit)).any(axis=1)]

# Check if the 'PUMP FAILURE' column (or the correct name) exists in the no_outliers dataframe
if 'PUMP FAILURE' in no_outliers.columns:

    # Separate the dataframe into failure and normal behavior
    failure_df = no_outliers[no_outliers['PUMP FAILURE (1 or 0)'] == 1]
    normal_df = no_outliers[no_outliers['PUMP FAILURE (1 or 0)'] == 0]

    # Create the box plots
    plt.figure(figsize=(12, 6))

    # Box plot for failure (PUMP FAILURE = 1)
    plt.subplot(1, 2, 1)
    sns.boxplot(data=failure_df.drop(columns=['PUMP FAILURE']))
    plt.title('Box Plot for PUMP FAILURE = 1 (Failure)')

    # Box plot for normal behavior (PUMP FAILURE = 0)
    plt.subplot(1, 2, 2)
    sns.boxplot(data=normal_df.drop(columns=['PUMP FAILURE']))
    plt.title('Box Plot for PUMP FAILURE = 0 (Normal Behaviour)')

    # Adjust layout and display plots
    plt.tight_layout()
    plt.show()
else:
    print("The column 'PUMP FAILURE' is not present in the dataframe.")

```

<b> i) Using the OLS Regression Model in the statsmodel.api library, create a regression equation that models the Pump Failure (Y-Variable) against all your independent variables in the dataframe_raw dataset. </b>

In order to fit a linear regression model with statsmodels.api there are a few steps that need to be taken. We have demonstrated this below:

<b> Don't forget to reimport the dataframes you've previously imported in Step 2 before starting these steps </b> 

    1. Establish two dataframes named, independent_variables and dependent_variables. The independent variables are known as explanatory variables - they help EXPLAIN what you are trying to model. Dependent Variable on the other hand is the variable of interest that you want to MODEL. In this case, the Dependent Variable is Pump Failure (1 or 0).
    
    2. Add a constant to your Independent Dataframe via the following syntax:
    independent_variables = sm.add_constant(independent_variables). This will simply add a constant stream of 1's in a column to your dataframe. This constant is used to account for bias in the model.  
    
    3. Store and Fit your model with the below syntax:
    regression_model = sm.OLS(dependent_variable,independent_variable).fit() 
    
    4. Print the regression_model.summary() to view the Regression Statistics 


<b> ii) Repeat i) but this time use the dataframe_stdev you imported previously.</b>

You will repeat the same steps as you have done in i) only you will be changing the dataset from dataset_raw to dataset_stdev. 


## Please put your code here 

```python





# Assume dataframe_raw is already imported.
# For example:
# dataframe_raw = pd.read_csv('your_data.csv')

# 1. Extract the dependent variable (PUMP FAILURE) and the independent variables (all others)
dependent_variable = dataframe_raw['PUMP FAILURE (1 or 0)']
independent_variables = dataframe_raw.drop(['PUMP FAILURE (1 or 0)'], axis=1)

# 2. Replace infinite values with NaN in both dependent and independent variables
dependent_variable.replace([np.inf, -np.inf], np.nan, inplace=True)
independent_variables.replace([np.inf, -np.inf], np.nan, inplace=True)

# *** Insert the pd.concat here ***
# Combine dependent and independent variables into a single DataFrame for cleaning:
combined = pd.concat([dependent_variable, independent_variables], axis=1)
print("Before cleaning, shape of combined:", combined.shape)

# Drop rows that contain any NaNs
combined.dropna(inplace=True)
print("After cleaning, shape of combined:", combined.shape)

# Reassign the cleaned data to dependent and independent variables
dependent_variable = combined['PUMP FAILURE (1 or 0)']
independent_variables = combined.drop(['PUMP FAILURE (1 or 0)'], axis=1)

# 3. Add a constant to the independent variables
independent_variables = sm.add_constant(independent_variables)

# 4. Fit the OLS regression model
regression_model = sm.OLS(dependent_variable, independent_variables).fit()

# 5. Print the regression summary
print(regression_model.summary())

```

```python


# ---------------------------------------------------
# Step 1: Generate Predictions
# ---------------------------------------------------
# We assume that `regression_model` is already defined from previous steps.
# For predictions, we must prepare the independent variables the same way as in training.

# In this example, we'll assume that the regression was built on all columns in dataframe_stdev
# except "PUMP FAILURE". (Adjust these steps if your model was built on a subset of columns.)
# Drop the dependent variable and add a constant column.
X_stdev = dataframe_stdev.drop(columns=['PUMP FAILURE'], errors='ignore')
X_stdev = sm.add_constant(X_stdev, has_constant='add')

# Generate predictions using the fitted model.
# This creates a new column "Prediction" in dataframe_stdev.
dataframe_stdev['Prediction'] = regression_model.predict(X_stdev)

# ---------------------------------------------------
# Step 2: Create a Dual-Axis Plot
# ---------------------------------------------------
# Axes One (primary) will plot the following variables:
#     "Volumetric Flow Meter 2", "Pump Efficiency", and "Horse Power".
# Axes Two (secondary) will plot "PUMP FAILURE" and "Prediction".

# Create a new figure and axis
fig, ax1 = plt.subplots(figsize=(12, 6))

# --- Primary Axis ---
# Plot the three variables on ax1.
ax1.plot(dataframe_stdev.index, dataframe_stdev['Volumetric Flow Meter 2'],
         label='Volumetric Flow Meter 2', color='blue', marker='o')
ax1.plot(dataframe_stdev.index, dataframe_stdev['Pump Efficiency'],
         label='Pump Efficiency', color='green', marker='o')
ax1.plot(dataframe_stdev.index, dataframe_stdev['Horse Power'],
         label='Horse Power', color='purple', marker='o')
ax1.set_ylabel("Flow & Efficiency Metrics", color='black')
ax1.legend(loc='upper left')
ax1.tick_params(axis='y', labelcolor='black')

# --- Secondary Axis ---
# Create a twin axis sharing the same x-axis.
ax2 = ax1.twinx()
# Plot Pump Failure and the Prediction on ax2.
ax2.plot(dataframe_stdev.index, dataframe_stdev['PUMP FAILURE'],
         label='Pump Failure', color='red', linestyle='--', marker='x')
ax2.plot(dataframe_stdev.index, dataframe_stdev['Prediction'],
         label='Prediction', color='orange', linestyle='--', marker='x')
ax2.set_ylabel("Pump Failure & Prediction", color='black')
ax2.legend(loc='upper right')
ax2.tick_params(axis='y', labelcolor='black')

# Add a title and show the plot.
plt.title("Dual-Axis Plot: Flow Metrics vs Pump Failure & Prediction")
plt.tight_layout()
plt.show()
```
